```python
# Podemos cargar un módulo o librería completa usando 'import'
import skimage

# La librería skimage está compuesta de diferentes submódulos, como data
# El siguiente comando nos permite descargar una imagen de corte histológico 
# incluído dentro de 'data'
ihc_rgb = data.immunohistochemistry()
print("ihc_rgb shape:", ihc_rgb.shape)
```

```python

# Para importar sólo el submódulo 'data' podemos utilizar '.' que en Python indica <<es parte de>>
import skimage.data
ihc_rgb = data.immunohistochemistry()
print("ihc_rgb shape:", ihc_rgb.shape)
```

```python
# Otra manera de importarlo es la siguiente
from skimage import data
ihc_rgb = data.immunohistochemistry()
print("ihc_rgb shape:", ihc_rgb.shape)
```

```python
# Incluso podemos renombrar el nombre del módulo importado
from skimage import data as dt
ihc_rgb = dt.immunohistochemistry()
print("ihc_rgb shape:", ihc_rgb.shape)
# O de otra forma
import skimage.data as md
ihc_rgb = md.immunohistochemistry()
print("ihc_rgb shape:", ihc_rgb.shape)
```

```python
# Esta es la manera común de trabajar con el módulo de gráficos de la librería matplotlib
import matplotlib.pyplot as plt
plt.imshow(ihc_rgb)
```

```python

# La librería externa pkgutil nos permite listar todos los módulos que componen 
# la librería sci-kit image
# https://docs.python.org/3/library/pkgutil.html
import pkgutil
import skimage
package=skimage
for importer, modname, ispkg in pkgutil.walk_packages(path=package.__path__,
                                                      prefix=package.__name__+'.',
                                                      onerror=lambda x: None):
    print(modname)
```
