## Librerías Python para Ciencia de Datos

### Numpy
Los arrays de NumPy son más rápidos y compactos que las listas de Python. Un array consume menos memoria y es conveniente de utilizar. NumPy utiliza mucho menos memoria para almacenar datos y proporciona un mecanismo para especificar los tipos de datos. Esto permite optimizar aún más el código.

* Más información sobre Numpy en [Numpy fundamentals](https://numpy.org/doc/stable/user/basics.html)

```python
from rich.console import Console
console = Console()
```

```python
console.rule("[cyan]Iniciar arrays numpy[/cyan]")

# Cargamos (importamos) el paquete numpy
import numpy as np

# Iniciación y operaciones con arrays:
a = np.arange(50)[0:20]
print("\n- a = ", a)

a = np.arange(1, 10, 2)
print(f"\n- a = {a}")

print("\n- Tipo de datos de a:", type(a))

l = [1, 2, 4, 6]
print("\n- Tipo de datos de l:", type(l))
print(l)

alist = np.array(l)
print("\n- alist:", alist)
```

```python
console.rule("[cyan]Creación de arrays[/cyan]")

# Array unidimensional
print("\n=== Array unidimensional ===")
x = np.linspace(2, 12, 34)
print(f"x = {x}\n")

# Otros ejemplos de creación de arrays de NumPy

# Array unidimensional con elementos específicos
print("=== Array unidimensional con elementos específicos ===")
y = np.array([1, 3, 5, 7, 9])
print(f"y = {y}\n")

# Array bidimensional con elementos aleatorios
print("=== Array bidimensional con elementos aleatorios ===")
z = np.random.rand(3, 4)
print(f"z = \n{z}\n")

# Array tridimensional con elementos constantes
print("=== Array tridimensional con elementos constantes ===")
w = np.full((2, 3, 2), 5)
print(f"w = \n{w}\n")

# Array de ceros
print("=== Array de ceros ===")
zeros = np.zeros((2, 5))
print(f"zeros = \n{zeros}\n")

# Array de unos
print("=== Array de unos ===")
ones = np.ones((3, 3))
print(f"ones = \n{ones}\n")
```

```python
console.rule("[cyan]Arrays de dimensión superior[/cyan]")

# Array de dimensión superior
print("\n=== Array de dimensión superior ===")
y = np.array([[1, 2.89, 3], [4.324353463, 5, 6], [1, 2, 5]], np.float32)
print(f"y = {y}\nShape: {y.shape}")

y = np.rint(y)
y = y.astype(np.int8)
print(f"\nConversión a enteros más próximos => y = {y}")

# Array bidimensional con elementos mixtos
print("\n=== Array bidimensional con elementos mixtos ===")
a = np.array([[1, 2, 2, 4, 3, 4], [1, 22.345245, 2, 42, 3, 4]])
z = np.unique(a)
print(f"Valores únicos en el array: {z}")
```

```python
console.rule("[cyan]Métodos varios a ejecutar sobre arrays[/cyan]")

print("Y:", y)

print("\nValor máximo:")
print("Max:", np.max(y))

print("\nValor mínimo:")
print("Min:", np.min(y))

print("\nMedia:")
print("Media:", np.mean(y))

print("\nDesviación estándar:")
print("Std:", np.std(y))

print("\nPercentil 25:")
print("Pct:", np.percentile(y, q=25))
```

```python
console.rule("[cyan]Slicing de arrays[/cyan]")
# Slicing con NumPy arrays

c = np.arange(1, 11, 1)
print("c:", c)

print("\nPrimeros 2 valores:")
print(c[:2])

print("\nÚltimos 8 elementos:")
print(c[2:])

print("\nÚltimos 2 valores:")
print(c[-2:])
```

## Numpy filtros

```python
console.rule("[cyan]Filtros sobre arrays (datos de imagen)[/cyan]")

import matplotlib.pyplot as plt
import numpy as np
from skimage import data, filters

# Cargar la imagen de muestra
image = data.camera()

# Crear una figura con subplots
fig, ax = plt.subplots(1, 3, figsize=(12, 4))

# Mostrar la imagen original
ax[0].imshow(image, cmap='gray')
ax[0].set_title('Imagen original')

# Aplicar filtro de desenfoque Gaussiano
blurred = filters.gaussian(image, sigma=2)
ax[1].imshow(blurred, cmap='gray')
ax[1].set_title('Filtro de desenfoque Gaussiano')

# Aplicar filtro Sobel para detectar bordes
edges = filters.sobel(blurred)
ax[2].imshow(edges, cmap='gray')
ax[2].set_title('Filtro Sobel para detección de bordes')

# Ajustar los espacios entre subplots
plt.tight_layout()

# Mostrar la figura
plt.show()
```

```python
console.rule("[cyan]Leer datos CSV con Pandas[/cyan]")

import pandas as pd

data_file = '/content/sample_data/california_housing_test.csv'

# Leer el archivo CSV y cargar los datos en un DataFrame
df = pd.read_csv(data_file)

# Mostrar las primeras filas del DataFrame
print("Primeras filas:\n", df.head())

# Mostrar las columnas del DataFrame
print("\nColumnas:\n", df.columns)

# Mostrar un rango de filas del DataFrame
print("\nRango de filas (457-462):\n", df.iloc[457:463])

# Mostrar información sobre el DataFrame
print("\nInformación:\n", df.info())

# Mostrar estadísticas descriptivas del DataFrame
print("\nDescripción:\n", df.describe())
```

```python
console.rule("[cyan]Otras operaciones básicas[/cyan]")

# Crear un DataFrame desde un diccionario
data = {'Nombre': ['Juan', 'María', 'Pedro', 'Ana'],
        'Edad': [25, 30, 40, 35],
        'Ciudad': ['Madrid', 'Barcelona', 'Valencia', 'Sevilla']}
df = pd.DataFrame(data)

# Mostrar las últimas filas del DataFrame
print("Últimas filas:\n", df.tail())

# Filtrar el DataFrame basado en una condición
filtered_df = df[df['Edad'] > 30]
print("\nFiltrado por edad:\n", filtered_df)

# Ordenar el DataFrame por una columna
sorted_df = df.sort_values('Edad', ascending=False)
print("\nOrdenado por edad:\n", sorted_df)

# Calcular la media de una columna
mean_age = df['Edad'].mean()
print("\nMedia de edad:", mean_age)

# Agrupar el DataFrame y calcular estadísticas por grupo
grouped_df = df.groupby('Ciudad').mean()
print("\nEstadísticas por ciudad:\n", grouped_df)

# Agregar una nueva columna calculada al DataFrame
df['Edad_Doble'] = df['Edad'] * 2
print("\nDataFrame con columna adicional:\n", df)

# Eliminar una columna del DataFrame
df = df.drop('Edad_Doble', axis=1)
print("\nDataFrame sin la columna Edad_Doble:\n", df)
```

```python
console.rule("[cyan]Gráficos con Seaborn[/cyan]")

import seaborn as sns
import matplotlib.pyplot as plt

# Cargar el conjunto de datos "iris"
iris = sns.load_dataset('iris')

# Gráfico de distribución (histograma y KDE)
sns.displot(iris['sepal_length'], kde=True)
plt.title('Distribución de la longitud del sépalo')
plt.xlabel('Longitud del sépalo')
plt.ylabel('Densidad')
plt.show()

# Gráfico de dispersión (scatter plot)
sns.scatterplot(x='sepal_length', y='sepal_width', hue='species', data=iris)
plt.title('Relación entre la longitud y el ancho del sépalo')
plt.xlabel('Longitud del sépalo')
plt.ylabel('Ancho del sépalo')
plt.show()

# Gráfico de barras (bar plot)
sns.barplot(x='species', y='petal_length', data=iris)
plt.title('Longitud promedio del pétalo por especie')
plt.xlabel('Especie')
plt.ylabel('Longitud promedio del pétalo')
plt.show()

# Gráfico de cajas (box plot)
sns.boxplot(x='species', y='petal_width', data=iris)
plt.title('Distribución del ancho del pétalo por especie')
plt.xlabel('Especie')
plt.ylabel('Ancho del pétalo')
plt.show()
```
